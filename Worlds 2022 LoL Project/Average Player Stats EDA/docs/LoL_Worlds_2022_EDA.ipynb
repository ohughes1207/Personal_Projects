```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
top_df = pd.read_csv('../data/clean/Top_df.csv')
jg_df = pd.read_csv('../data/clean/Jungle_df.csv')
mid_df = pd.read_csv('../data/clean/Mid_df.csv')
bot_df = pd.read_csv('../data/clean/Bot_df.csv')
supp_df = pd.read_csv('../data/clean/Support_df.csv')
```

```python
roles=['Top', 'Jungle', 'Mid', 'Bottom', 'Support']
df = pd.concat([top_df, jg_df, mid_df, bot_df, supp_df], keys=roles)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
df.loc[df['Team']=='T1']
```

```python
num = ['Games', 'Wins', 'Losses', 'Win Rate %', 'Kills', 'Deaths', 'Assists', 'KDA', 'Creep Score (CS)', 'CS per minute', 'Gold', 'Damage', 'Damage per minute', 'Kill Participation %', 'Kill Share %', 'Gold Share %', 'Unique Champions']
cat = ['1st Most Played', '2nd Most Played', '3rd Most Played']
```

```python
for role in roles:
    df_cat = df.loc[role][cat]
    for i in df_cat.columns:
            sns.barplot(x=df_cat[i].value_counts().index, y=df_cat[i].value_counts()).set(title=role+' '+i)
            plt.show()
```

```python
for role in roles:
    df_num = df.loc[role][num]
    for i in df_num.columns:
        plt.hist(df_num[i])
        plt.title(role+' '+i)
        plt.show()
```

```python
for role in roles:
    df_num = df.loc[role][num]
    sns.heatmap(df_num.corr(), cmap='jet').set(title=role+' Correlation Heat Map')
    plt.show()
```
